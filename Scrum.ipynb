# Méthode Agile Scrum

## Guide complet sur la méthode agile Scrum.

### 1. Qu’est ce que la méthode Scrum ? Quel est son objectif ?

La méthode Scrum est une méthode de gestion de projet agile inspirée du monde de rugby, basée sur une approche dynamique et participative de la conduite du projet, en se réunissant le plus souvent possible afin de vérifier l'avancer correcte du projet, toujours prêts à réorienter ce dernier au fil de son avancement. L'objectif est d’améliorer la productivité de l'équipe.
    

### 2. La méthode Scrum repose sur des valeurs, des principes et des piliers. Que sont-ils ? Expliquez les.
#### Les valeurs : 

- **Engagement** : En scrum, les équipes de développement et le product owner sont autonomes sur leurs activités respectives. Cela va amener ceux-ci à un engagement fort. Les équipes scrum vont s’engager à produire, à prendre leur autonomie, à prendre des décisions fortes sur le produit. Sans engagement, la mise en place de ce framework agile ne fonctionnera pas.

      
- **Focus** : Pour suivre un rythme soutenable et de qualité, les membres de l’équipe scrum devront être focus sur le produit qu’ils réalisent, sur les objectifs à atteindre. Il est essentiel d’être focus sur le produit si les équipes veulent trouver rapidement des résultats à la hauteur de leurs attentes. Parfois l’aspect focus est également nécessaire sur l’individus lui même ; en effet, il ne faut pas faire plusieurs choses à la fois mais privilégier de se focaliser sur une tâche précise et de bien la faire.

      
- **Ouverture** : Si scrum devient très populaire, très utilisé au sein des entreprises, il est parfois une  transcription d’un mini cycle en V. L’ensemble des membres de l’équipe doivent avoir une ouverture d’esprit et accepter les règles de ce framework agile qui sont loin d’être facile à suivre.

      
- **Respect** : Le scrum prône le respect au sein des équipes. C’est probablement la raison de dire qu’il n’y a pas de titre au sein des membres de l’équipe de développement. Le respect entre les différents membres de l’équipe de l’équipe sera essentiel pour avancer dans la même direction.

      
- **Courage** :Devenir autonome n’est pas si facile surtout dans certaines organisations complexes. Il faudra du courage aux équipes scrum pour atteindre cette autonomie. Il faudra également du courage au product owner pour savoir dire « NON » et rester le patron de son backlog et de sa priorisation.Les équipes comprennent très vite cette valeur essentielle des 5 valeurs scrum.

    
#### Les principes:

- Satisfaire le client en priorité
- Accueillir favorablement les demandes de changement
- Livrer le plus souvent possibles des versions opérationnelles de l’application
- Assurer une coopération permanente entre le client et l’équipe projet
- Construire des projets autour d’individus motivés
- Privilégier la conversation en face à face
- Mesurer l’avancement du projet en termes de fonctionnalités de l’application
- Faire avancer le projet à un rythme soutenable et constant
- Porter une attention continue à l’excellence technique et à la conception
- Faire simple
- Responsabiliser les équipes
- Ajuster à intervalles réguliers son comportement et ses processus pour être plus efficace
#### Les Piliers :

- Transparence
- Inspection
- Adaptation

### 3. Toute équipe de méthodes agiles est auto organisée. Comment se compose une équipe Scrum ? Expliquez leur rôle et leur importance.



#### Product Owner (PO)
- Garant de la vision du client.
- Le PO réalise des choix fonctionnels, en priorisant le développement d’une fonctionnalité au détriment d’une autre.
- Il constitue le Product Backlog qu’il rend transparent, visible et compris.
#### Scrum Master (SM)
- Facilitateur du projet et garant de la méthode Scrum.
- Il veille à conserver une harmonie au sein de l'équipe et à stimuler la créativité.
- Il est le garant de l'avancement du projet (outils de management visuel / événements timeboxés).
#### L’équipe de développement (Equipe de dev.)
- Elle va mettre en œuvre les solutions techniques, réaliser les développements.
- Idéalement, elle est stable et est composée de trois à dix personnes.
- Elle a le pouvoir de s’auto-gérer.
- Elle est pluridisciplinaire et possède toutes les compétences nécessaires pour créer un incrément du produit.

### 4. Un projet organisé par la méthode Scrum est rythmé par un ensemble d'événements. Énoncez ces différents événements et expliquez les.


#### Sprint Planning:
- Il s’agit du premier événement (cérémonie/rituel) qui débute une itération en Scrum. C’est une séance de discussions, d’écoute et de prises de décisions collectives.

#### Daily Scrum :
- Le Daily Scrum aussi connu sous le petit nom de Daily Stand Up Meeting, c'est une réunion quotidienne de 15 minutes. Chaque participant se tient généralement debout, devant le tableau d’avancement du projet afin de visualiser en temps réel les User Stories en cours de développement, en attente ou encore en cours de conception. L’objectif est de partager l’avancement des uns et des autres et de résoudre les points bloquants qui empêcheraient d’atteindre le Sprint Goal.
#### Sprint Review :
- La Sprint Review a pour objectif de faire un point sur ce qui a été réalisé en cours de sprint avec toutes les parties prenantes
#### Rétrospective :
- Suite à la review, l’équipe organise la dernière cérémonie du sprint avant de commencer l’itération suivante. Ce rituel offre à l’équipe un espace où les membres peuvent réfléchir librement au fonctionnement de l’équipe et à la manière de s’améliorer.
 - efficacité des processus ou des outils
 - relations entre les membres
 - problèmes intervenus dans le sprint
 - ce qui a bien fonctionné / ce qui n’a pas fonctionné
- L’objectif est de trouver un plan d’action pour le prochain sprint qui capitalise sur ces remarques.

### 6. Définissez de manière claire et concise, les mots clés suivants : user story, story points, la DoD, un Burndown chart et la méthode Kanban.
- **User story** : est une description de fonctionnalité écrite de façon à mettre en avant les besoins utilisateurs. Elle est écrite de manière compréhensible par l’ensemble des acteurs du projet ou liés à celui-ci.

- **Story points** :est comme une unité de mesure servant à estimer la charge de travail globale des équipes, afin de planifier au mieux chaque sprint ou itération en gestion de projet Agile Scrum.

- **DoD ou Definition of Done** : est un ensemble de critères qui doivent être remplis pour qu'un projet soit considéré comme « terminé ». Il s'agit essentiellement d'une liste de contrôle utilisée par les équipes Scrum pour créer une compréhension commune de ce qui est nécessaire pour rendre un produit livrable .

- **Burndown chart** :est un graphique de mesure  qui montre le travail achevé par jour par rapport au taux d'achèvement prévu. Son but est de permettre que le projet soit sur la bonne voie pour livrer la solution attendue dans les délais souhaités.

- **Méthode Kanban** : est un outil permettant de suivre l'évolution des différentes taches du projet
